```python
import os
import pandas as pd
import glob
from datetime import datetime
from bs4 import BeautifulSoup
import requests
import urllib

selection = 'marketinsights'
directory = "./utils/data/Scraped News/"
file_pattern = f"{selection}_data_*.csv"
files = glob.glob(os.path.join(directory, file_pattern))
print(files)
latest_file = max(files, key=os.path.getctime)
print(latest_file)

df = pd.read_csv(latest_file)
df.head()
```

```python
test_content = df.iloc[0]['Article content']
```

```python
from transformers import pipeline
model = 'distilbert/distilbert-base-uncased-finetuned-sst-2-english'
# create pipeline for sentiment analysis
classification = pipeline(model = model, task = 'sentiment-analysis')
classification(test_content)
```
